Bank Accounts according to Region

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x = 'Region',palette='Set1', data = df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Bank Accounts according to State

```python
plt.figure(figsize=(10,10))
ax = sns.countplot(x = 'State',palette='Set2', data = df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

We group the Population Group variable using State: This gives us the number of bank accounts according to the Population Group in a state wise group.

```python
facet = sns.FacetGrid(df, hue = 'State', size=11, aspect=1, palette='viridis')
facet.map(plt.hist, 'Population Group', bins=30, alpha=0.5)
facet.add_legend()
```

We group the Bank Group variable using State: This gives us the number of bank accounts according to the Bank Group in a state wise group.

```python
facet = sns.FacetGrid(df, hue ='State', size=11, aspect= 1, palette='viridis')
facet.map(plt.hist, 'Bank Group', bins=50, alpha=0.5)
facet.add_legend()

plt.tight_layout()
plt.show()
```
